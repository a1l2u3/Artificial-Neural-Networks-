1. Write a Python program to plot a few activation functions that are being used in neural networks.

Start

1. Define activation functions (e.g., sigmoid, relu, tanh)
2. Define input range
3. Calculate activation outputs for each function
4. Plot activation functions
5. Display plot

End


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    exp_values = np.exp(x - np.max(x, axis=0))
    return exp_values / np.sum(exp_values, axis=0)
```

```python
# Define input range
x = np.linspace(-5, 5, 100)
```

```python
# Calculate activation outputs
sigmoid_y = sigmoid(x)
relu_y = relu(x)
tanh_y = tanh(x)
```

```python
# Plot the activation functions
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, sigmoid_y, label='Sigmoid', color='b')
plt.title('Sigmoid Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
```

```python
plt.subplot(1, 3, 2)
plt.plot(x, relu_y, label='ReLU', color='r')
plt.title('ReLU Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
```

```python
plt.subplot(1, 3, 3)
plt.plot(x, tanh_y, label='Tanh', color='g')
plt.title('Tanh Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```
